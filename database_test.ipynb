```python
import pandas as pd
import json 

airlines = db["airlines"]
airports = db["airports"]
flights = db["flights"]

data_airlines = open('/home/houda/DST-airlines/2_datasets/airlines.json')
data_airlines = json.load(data_airlines)
airlines.insert_many(data_airlines)

data_airports = open('/home/houda/DST-airlines/2_datasets/airports.json')
data_airports = json.load(data_airports)
airports.insert_many(data_airports)

data_flights = open('/home/houda/DST-airlines/2_datasets/flights.json')
data_flights = json.load(data_flights)
flights.insert_many(data_flights)
```
